# UrbanGuard ML Models Presentation

This notebook summarizes the machine learning models used in UrbanGuard and provides small runnable demos for presenters.

## What the project uses
- **Road-risk model**: `RandomForestClassifier`
- **Pothole vibration model**: `RandomForestClassifier`

Both models are trained from `backend/road_data.csv` and saved as `.pkl` files with `joblib`.

## 1. Road-risk model

### Purpose
Predict whether a road location is **risky** or **safer**.

### Input features
- `lat`
- `lng`
- `confidence`
- `hour`
- `day`

### Target
The target is binary:
- `1` = risky road event (`pothole` or `crash`)
- `0` = safer road event

### Why Random Forest
Random Forest works well for this kind of tabular sensor data because it can capture non-linear patterns, is robust to noisy inputs, and performs well without heavy preprocessing.

## 2. Pothole vibration model

### Purpose
Detect pothole-like vibration patterns from device motion readings.

### Input features
- `ax`
- `ay`
- `az`
- `magnitude`
- `jerk`
- `intensity`

### Target
- `1` = pothole vibration detected
- `0` = no pothole detected

### Feature engineering
- `jerk` is computed from the change in magnitude
- `intensity` is computed as `ax + ay + az`

### Why Random Forest
It handles mixed signal patterns well, is easy to explain to non-technical audiences, and gives fast predictions in real time.

## 3. Model files used by the backend

The FastAPI backend loads these artifacts:
- `backend/road_risk_model.pkl`
- `backend/pothole_vibration_model.pkl`

The serving API exposes:
- `/predict-road-risk` for road classification
- `/detect-pothole` for vibration detection

```python
from pathlib import Path

import joblib
import pandas as pd

base_dir = Path.cwd()
if (base_dir / 'backend').exists():
    backend_dir = base_dir / 'backend'
else:
    backend_dir = base_dir

road_model_path = backend_dir / 'road_risk_model.pkl'
vibration_model_path = backend_dir / 'pothole_vibration_model.pkl'

road_model = joblib.load(road_model_path)
vibration_model = joblib.load(vibration_model_path)

print('Road-risk model type:', type(road_model).__name__)
print('Vibration model type:', type(vibration_model).__name__)
```

```python
road_sample = pd.DataFrame(
    [[12.9352, 77.6012, 0.91, 14, 20]],
    columns=['lat', 'lng', 'confidence', 'hour', 'day'],
)

road_prediction = road_model.predict(road_sample)[0]
print('Road sample:')
display(road_sample)
print('Prediction:', int(road_prediction))
print('Meaning:', 'Risky road' if road_prediction == 1 else 'Safer road')
```

```python
vibration_sample = pd.DataFrame(
    [[2.25, 2.05, 2.45, 3.81, 0.64, 6.75]],
    columns=['ax', 'ay', 'az', 'magnitude', 'jerk', 'intensity'],
)

vibration_prediction = vibration_model.predict(vibration_sample)[0]
print('Vibration sample:')
display(vibration_sample)
print('Prediction:', int(vibration_prediction))
print('Meaning:', 'Pothole vibration detected' if vibration_prediction == 1 else 'No pothole detected')
```

## 5. Presenter summary

UrbanGuard uses two lightweight, explainable tree-ensemble models:
1. A **road-risk classifier** that turns location and time features into a safe vs risky prediction.
2. A **vibration-based pothole detector** that turns IMU-style motion readings into a pothole/no-pothole prediction.

This setup is practical for real-time demos because the models are fast, easy to retrain, and simple to explain to non-technical audiences.
